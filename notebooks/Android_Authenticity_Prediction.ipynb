# Project Summary:

Android is the most in demand operating system in mobile manufacturers community. But the problem of malware apps is a concern for all the users as well. In the project I worked with the apps related dataset to classify the apps as malware or benign.
            First did the data cleaning by filling the null values and dropping the duplicates. 


Then explored the features and their relations with the labels.
Then, create the copy with only permissions indicator features, the category type and permission counts. Since, the category was a categorical column, used One-hot encoding to convert it into numerical feature.
           

For Data modelling, first did the train test split, after that first used tree-based methods as they don’t require scaling like boosting methods: Random-Forest, Gradient Boosting. After that scaled the data using Standard Scaler and applied K-Nearest neighbors, Support Vector Machines and Naïve Bayes and stored the Classification metrics in a dataframe and compared. After comparison we find that Gaussian Naïve Bayes was not working well. We did hyperparameter tuning on the KNN, Random Forest and Gradient Boosting since they were doing quite well in all the metrics. After comparing the scores, I decided to go and create a final model with Gradient boosting and then fit the entire data to the final model.

# Null Values Analysis:

```python
# Null values analysing and treating them: 
df.isnull().sum().sort_values(ascending = False)
```

```python
# column names with null values:
null_columns = [col for col in df.columns if df[col].isnull().sum()]
print(null_columns)
```

```python
# Dropping the row with no app name:-- won't affect data that much as effective row is only 1
df = df.dropna(subset = ['App'])
```

```python
# Treating null values of description column:
df[df['Description'].isnull()]
```

```python
df['Description'] = df['Description'].fillna('')
```

```python
# List of Dangerous permissions and safe permissions:
danger_perm = [col for col in df.columns if col[-3:]=='(D)']
safe_perm = [col for col in df.columns if col[-3:]=='(S)']
```

```python
# Function to find Dangerous permission count to fill the null values:
def danger_perm_count(x):
    global danger_perm
    return float(x[danger_perm].sum())
```

```python
df.loc[df['Dangerous permissions count'].isnull(),['Dangerous permissions count']]= df.loc[df['Dangerous permissions count'].isnull()].apply(danger_perm_count, axis=1)
```

```python
# Filling null values of related apps as empty string since can't extrapolate them:
df['Related apps'] = df['Related apps'].fillna('')
```

```python
df.isnull().sum().sort_values(ascending = False)
```

# Function for reporting the evaluation metrics for the model:

```python
# Reporting Model:
def report_model(model):
  
  # Fitting & predicting the data :
  model.fit(X_train, y_train.ravel())
  train_pred = model.predict(X_train)
  test_pred = model.predict(X_test)

  # Train Data Metrics : 
  print("Train Results are:")
  global train_accuracy, train_precision, train_recall, train_roc_auc
  print(classification_report(y_train,train_pred))
  print("\n \n")
  print("Confusion Matrix is:")
  print(confusion_matrix(y_train,train_pred))
  print("\n \n")
  print("ROc-AUC Score is:")
  print(roc_auc_score(y_train,train_pred))

  # Appending the metrics to the global list for further comparing various models:
  train_accuracy.append(accuracy_score(y_train, train_pred))
  train_precision.append(precision_score(y_train, train_pred))
  train_recall.append(recall_score(y_train, train_pred))
  train_roc_auc.append(roc_auc_score(y_train, train_pred))
  print("\n\n\n\n\n")
  print(" Test Results are:")

  # Test Data Metrics:
  global test_accuracy, test_precision, test_recall, test_roc_auc
  print(classification_report(y_test,test_pred))
  print("\n \n")
  print("Confusion Matrix is:")
  print(confusion_matrix(y_test,test_pred))
  print("\n \n")
  print("ROc-AUC Score is:")
  print(roc_auc_score(y_test,test_pred))

  # Appending the metrics to the global list for further comparing various models:
  test_accuracy.append(accuracy_score(y_test, test_pred))
  test_precision.append(precision_score(y_test, test_pred))
  test_recall.append(recall_score(y_test, test_pred))
  test_roc_auc.append(roc_auc_score(y_test, test_pred))
```

```python
# Lists for storing metrics:
train_accuracy= []
train_precision = []
train_recall = []
train_roc_auc = []
test_accuracy= []
test_precision = []
test_recall = []
test_roc_auc = []
```
